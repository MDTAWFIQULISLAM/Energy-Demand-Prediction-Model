```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



data = {
    'Algorithm': ['fbprophet', 'fbprophet', 'fbprophet', 'CatBoost', 'CatBoost', 'CatBoost', 'neural_prophet', 'neural_prophet', 'neural_prophet'],
    'Train_Test': ['train', 'validation', 'test', 'train', 'validation', 'test', 'train', 'validation', 'test'],
    'MAE': [939.3951, 1723.264, 1283.042, 1244.502, 1418.731, 4527.32, 798.5524, 1458.2922, 1580.2788],
    'RMSE': [1335.864, 2198.676, 1745.709, 1708.351, 1949.408, 5245.25, 1277.8469, 2009.986, 2310.1311],
    'MAPE': [3.8406, 7.9468, 5.8317, 5.1612, 6.3079, 23.2534, 3.2255, 6.6024, 6.8866],
    'r2_score': [94.9816, 77.5245, 86.8772, 91.7928, 82.3318, -18.471, 95.4038, 81.2166, 77.0197]
}

# Adding the 'Difference' data as a separate DataFrame for 'test' phase.
difference_data = {
    'Algorithm': ['fbprophet', 'CatBoost', 'neural_prophet'],
    'Difference': [343.647, 3282.818, 781.7264]
}

df = pd.DataFrame(data)
df_difference = pd.DataFrame(difference_data)

# Set up the matplotlib figure
fig, axes = plt.subplots(2, 2, figsize=(18, 10))

# Bar plot for MAE by algorithm and phase
sns.barplot(ax=axes[0, 0], x='Algorithm', y='MAE', hue='Train_Test', data=df)
axes[0, 0].set_title('Mean Absolute Error (MAE) by Algorithm and Phase')

# Line plot for r2 score by algorithm and phase
sns.lineplot(ax=axes[0, 1], x='Algorithm', y='r2_score', hue='Train_Test', data=df, marker='o')
axes[0, 1].set_title('R^2 Score by Algorithm and Phase')

# Bar plot for RMSE by algorithm and phase
sns.barplot(ax=axes[1, 0], x='Algorithm', y='RMSE', hue='Train_Test', data=df)
axes[1, 0].set_title('Root Mean Squared Error (RMSE) by Algorithm and Phase')

# Bar plot for MAE differences in the 'test' phase
sns.barplot(ax=axes[1, 1], x='Algorithm', y='Difference', data=df_difference, palette='Set2')
axes[1, 1].set_title('Difference in MAE between Train and Test (Test Phase)')

plt.tight_layout()
plt.show()
```
